What is Learning(training)?  
: **find w that minimize the loss**  
: `arg min loss(w)`

## **Implementation**

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0

def forward(x):
    return x * w

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y)**2

def gradient(x, y):
    return 2 * x * (x * w - y)

print("predict before training : ", 4, forward(4))

# training
for epoch in range(100):
    for x_val, y_val in zip(x_data, y_data):
        grad = gradient(x_val, y_val)
        w = w - 0.01 * grad
        print("\tgrad: ", x_val, y_val, grad)
        l = loss(x_val, y_val)
    print("progress: ", epoch, "w=", w, "loss= ", l)

print("predict after training: ", forward(4))
```

# My Implementation

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0

def y_hat(x):
    return x * w

def loss(x, y):
    y_pred = y_hat(x)
    return (y_pred - y)**2

def gradient(x, y):
    return 2 * x * (x * w - y) # loss를 w에 대해 편미분 한 후의 값

print("predict before training", y_hat(4))

for epoch in range(100):
    for x_val, y_val in zip(x_data, y_data):
        grad = gradient(x_val, y_val)
        w = w - 0.01 * grad
        print("\tgrad: ", x_val, y_val, grad)
        l = loss(x_val, y_val)
    print("progress: ", epoch, "w = ", w, "loss = ", l)

print("predict after training", y_hat(4))
```
