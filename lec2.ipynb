y hat = x * w + b  
(w = weight, b = bias)  
simplied) y hat = x * w

결국은 우리 prediction을 TRUE LINE에 가깝게 만드는게 목표

`loss = (y hat - y)**2`  
     = `(x * w - y)**2`  
-> loss(error)를 최소화하는게 목표  
-> w : random guess  
-> loss 간의 mean 계산   
    = MSE(Mean Squared Error) = `1/N * sigma(yn hat - yn)**2`


**실습**

```python
import numpy as np
import matplotlib.pyplot as plt # 시각화

x_data = [1.0, 2.0, 3.0] # input
y_data = [2.0, 4.0, 6.0] # output

w = 1.0 # random guess

def forward(x):
    return x * w 

# loss function
def loss(x, y): 
    y_pred = forward(x) # y hat(우리의 예측값)
    return (y_pred - y)**2

w_list = []
mse_list = []

for w in np.arange(0.0, 4.1, 0.1): # 0.0부터 4.1까지 0.1단위로 반환
    print("w = ", w)
    l_sum = 0 ## MSE를 구하기 위한 total
    for x_val, y_val in zip(x_data, y_data):
        y_pred_val = forward(x_val)
        l = loss(x_val, y_val)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    print("MSE = ", l_sum / 3)
    w_list.append(w)
    mse_list.append(l_sum / 3)

plt.plot(w_list, mse_list)
plt.ylabel("Loss")
plt.xlabel("w")
plt.show()
```

MSE를 나타내는 loss function을 만들고 이를 시각화하는 작업까지  
: 내가 직접 복습한 코드

```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w_list = []
mse_list = []
w = 1.0 # random guess

def y_hat(x):
     return w * x

def loss(x, y):
     y_pred = y_hat(x)
     return (y_pred - y)**2

for w in np.arange(0.0, 4.1, 0.1):
    print("w = ", w)
    loss_sum = 0 # loss의 총합
    for x_val, y_val in zip(x_data, y_data):
        y_pred_val = y_hat(x_val)
        l = loss(x_val, y_val)
        loss_sum += l # 각각 loss를 구해서 loss_sum에다가 합치기
        print(f"\t x_val : {x_val}, y_val : {y_val}, y_pred_val : {y_pred_val}, loss : {l}")

    print("MSE = ", loss_sum / 3)
# 그래프를 만들기 위한 과정
    w_list.append(w)
    mse_list.append(loss_sum / 3)

plt.plot(w_list, mse_list)
plt.ylabel("mse")
plt.xlabel("w")
plt.show()
```
